```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt

# Define parameters
means = [0, 0, 0]
variances = [1, 2, 5]
x_values = np.linspace(-10, 10, 400)

# Plot different normal distributions
plt.figure(figsize=(8, 5))
for mean, variance in zip(means, variances):
    y_values = [normal_density(mean, variance, x) for x in x_values]
    plt.plot(x_values, y_values, label=f"μ={mean}, σ²={variance}")

plt.title("Normal Distributions")
plt.xlabel("x")
plt.ylabel("Density f_X(x)")
plt.legend()
plt.grid()
plt.show()


def normal_density(mean, variance, x):
    sigma = np.sqrt(variance)  # Standard deviation is sqrt(variance)
    coefficient = 1 / (np.sqrt(2 * np.pi * variance))
    exponent = np.exp(-((x - mean) ** 2) / (2 * variance))
    return coefficient * exponent

def midpoint_integration(mean, variance, a, b, n=1000):
    dx = (b - a) / n  # Step size
    total = 0  # Accumulator for the sum

    for i in range(n):
        mid_x = a + (i + 0.5) * dx  # Midpoint of the subinterval
        total += normal_density(mean, variance, mid_x) * dx  # Approximate area

    return total

mean = 171
variance = 7.12
a, b = 162, 190

probability = midpoint_integration(mean, variance, a, b, n=1000)
print(f"Probability of height between {a}cm and {b}cm: {probability:.4f}")
```

```python
import numpy as np

def expected_dosage(mean, variance, coefficient=2.38):
    E_X2 = variance + mean**2  # Compute E[X^2]
    return coefficient * E_X2  # Compute E[D(X)]

# Given parameters
mean_height = 171
variance_height = 7.1

dosage_expectation = expected_dosage(mean_height, variance_height)
print(f"Estimated average dosage: {dosage_expectation:.2f}")

# Part (b) - Expected value of an exponential distribution
def expected_exponential(lambda_rate):
    return 1 / lambda_rate

lambda_pandemics = 1/50
expected_pandemic_time = expected_exponential(lambda_pandemics)
print(f"Expected time between pandemics: {expected_pandemic_time} years")

# Part (a) - Expected value of a uniform distribution
def expected_uniform(a, b):
    return (a + b) / 2

a, b = 0, 10  # Example range
expected_uniform_value = expected_uniform(a, b)
print(f"Expected value for uniform distribution on [{a}, {b}]: {expected_uniform_value}")
```
