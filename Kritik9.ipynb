```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Exercise (a): Partial derivatives of f(x,y) = e^x sin(y) + y^3
def exercise_a():
    print("Exercise (a):")
    x, y = sp.symbols('x y')
    f = sp.exp(x) * sp.sin(y) + y**3

    # First partial derivatives
    df_dx = sp.diff(f, x)
    df_dy = sp.diff(f, y)

    print("∂f/∂x =", df_dx)
    print("∂f/∂y =", df_dy)
    print()
    return x, y

# Exercise (b): Gradient of g(x,y) = x^2y + xy^2 at (1,-1)
def exercise_b():
    print("Exercise (b):")
    x, y = sp.symbols('x y')
    g = x**2 * y + x * y**2

    # Gradient vector
    grad_g = [sp.diff(g, var) for var in (x, y)]
    print("∇g =", grad_g)

    # Magnitude at (1, -1)
    grad_magnitude = sp.sqrt(sum(comp**2 for comp in grad_g))
    magnitude_at_point = grad_magnitude.subs({x: 1, y: -1})
    print("Magnitude of ∇g at (1, -1) =", magnitude_at_point)
    print()
    return x, y

# Exercise (c): Second partial derivatives of h(x,y) = ln(x^2 + y^2)
def exercise_c():
    print("Exercise (c):")
    x, y = sp.symbols('x y')
    h = sp.log(x**2 + y**2)

    # Second partial derivatives
    d2h_dx2 = sp.diff(h, x, x)
    d2h_dy2 = sp.diff(h, y, y)
    d2h_dxdy = sp.diff(h, x, y)

    print("∂²h/∂x² =", d2h_dx2)
    print("∂²h/∂y² =", d2h_dy2)
    print("∂²h/∂x∂y =", d2h_dxdy)

    # Verify symmetry of mixed partials
    d2h_dydx = sp.diff(h, y, x)
    print("\nSymmetry check (∂²h/∂x∂y == ∂²h/∂y∂x):",
          sp.simplify(d2h_dxdy - d2h_dydx) == 0)
    print("This demonstrates Clairaut's theorem on equality of mixed partials.")
    print()
    return x, y

# Exercise (d): Contour plot of j(x,y) = x^3 - 3xy + y^3
def exercise_d():
    print("Exercise (d): Plotting j(x,y) = x^3 - 3xy + y^3")
    x, y = sp.symbols('x y')
    j = x**3 - 3*x*y + y**3

    # Convert to numerical function
    j_func = sp.lambdify((x, y), j, 'numpy')

    # Create grid and evaluate
    x_vals = np.linspace(-3, 3, 400)
    y_vals = np.linspace(-3, 3, 400)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = j_func(X, Y)

    # Plot
    plt.figure(figsize=(10, 8))
    plt.contourf(X, Y, Z, levels=50, cmap='viridis')
    plt.colorbar()
    plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.show()
    print()
    return x, y

# Exercise (e): Contour plot of a custom function
def exercise_e():
    print("Exercise (e): Plotting custom function k(x,y) = sin(x)cos(y)exp(-(x^2+y^2)/10)")
    x, y = sp.symbols('x y')
    k = sp.sin(x) * sp.cos(y) * sp.exp(-(x**2 + y**2)/10)

    # Convert to numerical function
    k_func = sp.lambdify((x, y), k, 'numpy')

    # Create grid and evaluate
    x_vals = np.linspace(-5, 5, 400)
    y_vals = np.linspace(-5, 5, 400)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = k_func(X, Y)

    # Plot
    plt.figure(figsize=(10, 8))
    plt.contourf(X, Y, Z, levels=50, cmap='plasma')
    plt.colorbar()
    plt.title('Contour plot of $k(x, y) = \sin(x)\cos(y)e^{-(x^2+y^2)/10}$')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.show()

# Run all exercises
if __name__ == "__main__":
    _ = exercise_a()
    _ = exercise_b()
    _ = exercise_c()
    _ = exercise_d()
    exercise_e()
```
